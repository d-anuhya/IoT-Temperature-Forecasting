## Datetime information
>datetime column has a lot of information such as year, month, weekday and so on. To utilize these information in EDA and modeling phase, we need extract them from datetime column.

```python
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y %H:%M:%S')
df['year'] = df['date'].apply(lambda x : x.year)
df['month'] = df['date'].apply(lambda x : x.month)
df['day'] = df['date'].apply(lambda x : x.day)
df['weekday'] = df['date'].apply(lambda x : x.day_name())
df['weekofyear'] = df['date'].apply(lambda x : x.weekofyear)
df['hour'] = df['date'].apply(lambda x : x.hour)
df['minute'] = df['date'].apply(lambda x : x.minute)
df.head(3)
```

## Seasonal information
><div class="alert alert-success" role="alert">
><ul>
>    <li><b>Winter</b> : December to February</li>
>    <li><b>Summer</b> : March to May</li>
>    <li><b>Monsoon</b> : June to September</li>
>    <li><b>Post-monsoon</b> : October to November</li>
></ul>
></div>

>function to convert month variable into seasons

```python
def month2seasons(x):
    if x in [12, 1, 2]:
        season = 'Winter'
    elif x in [3, 4, 5]:
        season = 'Summer'
    elif x in [6, 7, 8, 9]:
        season = 'Monsoon'
    elif x in [10, 11]:
        season = 'Post_Monsoon'
    return season
```

```python
df['season'] = df['month'].apply(month2seasons)
df.head(3)
```

## Timing information
><div class="alert alert-success" role="alert">
><ul>
>    <li><b>Night</b> : 22:00 - 23:59 / 00:00 - 03:59</li>
>    <li><b>Morning</b> : 04:00 - 11:59</li>
>    <li><b>Afternoon</b> : 12:00 - 16:59</li>
>    <li><b>Evening</b> : 17:00 - 21:59</li>
></ul>
></div>

```python
def hours2timing(x):
    if x in [22,23,0,1,2,3]:
        timing = 'Night'
    elif x in range(4, 12):
        timing = 'Morning'
    elif x in range(12, 17):
        timing = 'Afternoon'
    elif x in range(17, 22):
        timing = 'Evening'
    else:
        timing = 'X'
    return timing
```

```python
df['timing'] = df['hour'].apply(hours2timing)
df.head(3)
```

```python
plt.figure(figsize=(6, 3))

# Round the date to the month level and count the frequency of each month
seconds_rd = np.round(df['date'].apply(lambda x: x.strftime("%Y-%m")).value_counts(normalize=True).sort_index() * 100, decimals=1)

# Bar graph
plt.bar(seconds_rd.index, seconds_rd.values)

plt.title('Monthly Readings Count', fontsize=14)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Percentage', fontsize=14)

plt.xticks(rotation=90, ha='center', fontsize=12)
plt.yticks(fontsize=12)

plt.show()

# Displaying the details
# display(seconds_rd)
```

### Temperature
>Temperature clearly consists of multiple distributions.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

print(df.columns)

# Calculate the histogram
hist, edges = np.histogram(df['temp'], density=True, bins=30)

# Calculate the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, df['temp'].mean(), df['temp'].std())

# Plot the histogram and normal distribution curve
plt.hist(df['temp'], bins=edges, alpha=0.3, color='green', label='Histogram', density=True)
plt.plot(x, p, 'k--', label='Normal Distribution')

# Set plot options
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Density')
plt.legend(loc='best')

# Display the plot
plt.show()
```
